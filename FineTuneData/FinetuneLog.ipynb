```python
import pandas as pd

# Les CSV-filen med korrekt separator og encoding
df = pd.read_csv("/content/norges_bank_1980-1999Finetune.csv", sep=";", encoding="ascii")

# Konverter dato-kolonnen
df['TIME_PERIOD'] = pd.to_datetime(df['TIME_PERIOD'], format='%d.%m.%Y')

# Sorter for sikkerhets skyld
df = df.sort_values('TIME_PERIOD').reset_index(drop=True)

# Konverter alle valutakolonner til numerisk (tvinger ugyldige til NaN)
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Fyll helgedager og eventuelle manglende verdier med forward fill
df = df.ffill()

# Valgfritt: sett TIME_PERIOD som indeks
df.set_index('TIME_PERIOD', inplace=True)

# Resultat
print(f"Antall observasjoner etter ffill: {len(df)}")
display(df.head(10))
```

```python
import pandas as pd
import numpy as np

# Les og klargjør data
df = pd.read_csv("/content/norges_bank_1980-1999Finetune.csv", sep=";", encoding="ascii")
df['TIME_PERIOD'] = pd.to_datetime(df['TIME_PERIOD'], format='%d.%m.%Y')
df = df.sort_values('TIME_PERIOD').reset_index(drop=True)

# Konverter kolonner til numerisk
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Fyll manglende verdier (typisk helger)
df = df.ffill()

# Log-transformer alle valutakolonner
df_log = df.copy()
df_log[df.columns[1:]] = np.log(df[df.columns[1:]])

# Sett dato som indeks
df_log.set_index('TIME_PERIOD', inplace=True)

# Vis resultat
print(f"Antall observasjoner etter log-transformasjon: {len(df_log)}")
display(df_log.head(10))
```

```python
# Lagre log-transformert datasett som CSV
output_path = "/content/norges_bank_1980-1999Finetune_log.csv"
df_log.to_csv(output_path)

print(f"Filen er lagret som: {output_path}")
```
